# Read the text file and print it

```python
import time

import spacy

# Load the model
nlp = spacy.load("en_core_web_sm")

start_time = time.time()

with open("./alice29.txt") as file:
    text = file.read()

print(text)
```

# Clean the text and write it to a new file

```python
doc = nlp(text)
cleaned_tokens = [];

for token in doc:
    # Filter unwanted tokens
    if (not token.is_punct and not token.is_space
            and not token.is_stop and not token.is_quote and token.is_ascii):
        cleaned_tokens.append(token.text.replace("--", "").removesuffix("-").removeprefix("-").lower())

with open('./alice29_cleaned.txt', 'w') as file:
    file.write(' '.join(cleaned_tokens))

with open('./alice29_words.txt', 'w') as file:
    file.write('\n'.join(cleaned_tokens))

```

```python
i = 1
with open('./alice29_sentences.txt', 'w') as file:
    for sentence in doc.sents:
        print(f"Sentence {i}: {sentence}\n")
        file.write(f"Sentence {i}: {sentence}\n")
        i += 1
```

# Top 50 Words in the Text

```python
from collections import Counter

counts = Counter(cleaned_tokens)

print("Top 10 words in the text:")
with open('./alice29_top10words.txt', 'w') as file:
    file.write("Top 10 words in the text:\n")
    # Calculate and print top ten words
    for word, count in counts.most_common(10):
        print(f"{word}: {count}")
        file.write(f"{word}: {count}\n")
```

# Calculate Execution Time

```python
stop_time = time.time()

print(f"Execution time: {stop_time - start_time} seconds")

with open('./alice29_time.txt', 'w') as file:
    file.write(f"Execution time: {stop_time - start_time} seconds")
```
